```python
directory='peter'
parent_dir='/content/drive/MyDrive'
path=parent_dir+'/'+directory
path
```

```python
import os
os.chdir(path)
os.getcwd()
```

```python
import pathlib
path=pathlib.Path(path)
file_list=[str(path) for path in path.glob('*.jpg')]
```

```python
print(len(file_list))
```

we can now visualise the images in the file_list as follows

```python
%matplotlib inline
import tensorflow as tf
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,8))
for i ,file in enumerate(file_list[:6]):
  img_raw=tf.io.read_file(file)
  img=tf.image.decode_jpeg(img_raw)
  ax=fig.add_subplot(3,3,i+1)
  ax.imshow(img)
  ax.set_xticks([])
  ax.set_yticks([])
  ax.set_title(file)

_=plt.tight_layout()
_=plt.show()
```
